```python
import pandas as pd
from pybaseball import statcast
```

```python
drop_cols = ['release_pos_x', 'release_pos_z', 'spin_dir','spin_rate_deprecated', 
             'break_angle_deprecated','break_length_deprecated', 'des', 'game_type',
             'game_year', 'pfx_x', 'pfx_z', 'plate_x', 'plate_z', 'hc_x', 'hc_y',
             'tfs_deprecated', 'tfs_zulu_deprecated', 'umpire', 'sv_id', 'vx0',
             'vy0', 'vz0', 'ax', 'ay', 'az', 'sz_top', 'sz_bot', 'launch_speed', 
             'launch_angle', 'release_pos_y', 'launch_speed_angle']
```

```python
#2017
df = statcast('2017-04-02', '2017-10-02')
df = df.drop(columns=drop_cols)

df.to_pickle(path='pitches_2017.pkl', compression='zip')
```

```python
#2018

df = statcast('2018-03-29', '2018-10-01')
df = df.drop(columns=drop_cols)

df.to_pickle(path='pitches_2018.pkl', compression='zip')
```

```python
#2019 (opening day - all star break)

df = statcast('2019-03-28', '2019-07-07')
df = df.drop(columns=drop_cols)

df.to_pickle(path='pitches_2019_first_half.pkl', compression='zip')
```
